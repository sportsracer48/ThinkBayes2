```python
# Configure Jupyter so figures appear in the notebook
%matplotlib inline

# Configure Jupyter to display the assigned value after an assignment
%config InteractiveShell.ast_node_interactivity='last_expr_or_assign'

# import classes from thinkbayes2
from thinkbayes2 import Pmf, Suite, Joint, MakePoissonPmf, MakeNormalPmf, MakeMixture

import numpy as np

import thinkplot

from math import exp,sqrt

lam = 1
```

```python
class Battleship(Suite, Joint):
    
    def Likelihood(self, data, hypo):
        """Computes the likelihood of the data under the hypothesis.
        
        data: tuple of ((x,y),true if hit, false if miss)
        hypo: tuple of (x,y) for ship location
        """
        guess,hit=data
        
        guess = np.array(guess)
        hypo  = np.array(hypo)
        d = np.linalg.norm(guess-hypo)
        if(hit):
            return exp(-lam*d)
        else:
            return 1-exp(-lam*d)
```

```python
W=10
H=10
prior = Battleship([(x,y) for x in range(W) for y in range(H)]);
prior.Update(((4,4),False))

thinkplot.Contour(prior)
```
